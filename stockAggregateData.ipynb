```python
stockTickers = pd.read_json("stockTickers_31032024.json")
stockTickers.info()
```

```python
polygonAPIkey = "r9VpJsap3oKUiXK1s6ae9PsI9jq18OBW"
dateFrom = '2022-04-01'
dateTo = '2023-03-28'

dailyStockTickerAggregates = []

for index, ticker in stockTickers.iterrows(): 
    
    tickerId = ticker['id']
    tickerName = ticker['ticker']
    
    tickerUrl = 'https://api.polygon.io/v2/aggs/ticker/{0}/range/1/day/{1}/{2}/?adjusted=true&sort=asc&limit=120&apiKey={3}'.format(tickerName, dateFrom, dateTo, polygonAPIkey)
    
    resp = urlopen(tickerUrl)
    print('fetching ticker aggs')
    time.sleep(12)
    aggsDataForTicker = json.loads(resp.read())
    
    
    if 'results' in aggsDataForTicker:
        for dailyTickerAgg in aggsDataForTicker['results']:

            dailyAgg = dailyTickerAgg
            dailyAgg['ticker_id'] = tickerId
            dailyAgg['date'] = dt.datetime.fromtimestamp(dailyTickerAgg['t'] / 1e3, tz=dt.timezone.utc).strftime("%Y-%m-%d %H:%M:%S")
            dailyAgg['name'] = tickerName
            
            dailyStockTickerAggregates.append(dailyAgg)
            
print('done fetching data')
```

```python
dailyStockTickerAggregates = pd.DataFrame(dailyStockTickerAggregates)
dailyStockTickerAggregates.info()
```
